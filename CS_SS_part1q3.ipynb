```python
from sklearn.model_selection import train_test_split
X = list(data.columns)
X.remove('Classification')
X.remove('App')
Xdata=data[X]
Ydata=data['Classification']
X_train, X_test, y_train, y_test = train_test_split(Xdata, Ydata, test_size=0.2) 
```
